## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
import matplotlib.image as mpimg
from scipy.fft import fft, ifft, fft2, ifft2
#from Contents.FFT import FFT, IFFT, FFT2D, IFFT2D
```

## Create a signal

```python
x = np.linspace(0, 6*2*np.pi, num = 1000)
signal_1 = np.sin(2*x)
signal_2 = np.cos(3*x)
signal = signal_1 + signal_2
```

## Plotting signal

```python
fig, ax = plt.subplots(figsize = (16,6))

ax.plot(x,signal)

ax.set_xlim(0, 6*2*np.pi);
```

```python
def FFT(signal):

    N = len(signal)
    if(N == 1):
        return signal + 0j
    
    n = np.arange(N)
    w = np.exp(2j*np.pi*n/N)
    
    signal_even, signal_odd = signal[::2], signal[1::2]
    y_even, y_odd = FFT(signal_even), FFT(signal_odd)

    y = np.zeros_like(signal, dtype = np.cdouble)
    
    for i in range(int(N/2)):
        t = w[i]*y_odd[i]
        y[i] = y_even[i] + t
        y[i+int(N/2)] = y_even[i] - t

    return y

def IFFT(signal):
    
    N = len(signal)
    if(N == 1):
        return signal + 0j
    
    n = np.arange(N)
    w = (1/N)*np.exp(-2j*np.pi*n/N)
    
    signal_even, signal_odd = signal[::2], signal[1::2]
    y_even, y_odd = FFT(signal_even), FFT(signal_odd)

    y = np.zeros_like(signal, dtype = np.cdouble)
    
    for i in range(int(N/2)):
        t = w[i]*y_odd[i]
        y[i] = y_even[i] + t
        y[i+int(N/2)] = y_even[i] - t
    
    return y
```

## FFT in signal

```python
signal_frequency = FFT(signal)
```

## Plotting fequency of signal

```python
fig, ax = plt.subplots(figsize = (16,6))

ax.plot(abs(signal_frequency))

ax.set_xlim(0, 1000);
```

## FFT inverse in signal

```python
signal_again = IFFT(signal_frequency)
```

## Plotting the reconstruct signal

```python
fig, ax = plt.subplots(figsize = (16,6))

ax.plot(x,abs(signal_again))

ax.set_xlim(0, 6*2*np.pi);
```

## Read image

```python
img = mpimg.imread('Images/cameraman.tif')
```

## Plotting image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

```python
def FFT2D(img):

    M, N = img.shape
    
    img_temp = np.zeros_like(img, dtype = np.cdouble)
    img_frequency = np.zeros_like(img, dtype = np.cdouble)

    for i in range(M):
        img_temp[i,:] = FFT(img[i,:])

    for i in range(N):
        img_frequency[:,i] = FFT(img_temp[:,i])

    return img_frequency

def IFFT2D(img_frequency):

    M, N = img_frequency.shape
    
    img_temp = np.zeros_like(img_frequency, dtype = np.cdouble)
    img = np.zeros_like(img_frequency, dtype = np.cdouble)

    for i in range(M):
        img_temp[i,:] = IFFT(img_frequency[i,:])

    for i in range(N):
        img[:,i] = IFFT(img_temp[:,i])

    return img
```

## FFT image

```python
img_FFT = FFT2D(img)
```

## Plotting Frequency domain image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(np.power(abs(img_FFT), 1/4), cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## FFT image inverse

```python
img_again = IFFT2D(img_FFT)
```

## Plotting image reconstruct

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(abs(img_again), cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```
