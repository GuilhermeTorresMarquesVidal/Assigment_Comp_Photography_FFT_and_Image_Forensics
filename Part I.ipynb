## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
import matplotlib.image as mpimg
from scipy.fft import fft, ifft, fft2, ifft2
from Contents.FFT import FFT, IFFT, FFT2D, IFFT2D
from Contents.FFT_old import fft2d, ifft2d
```

## Create a signal

```python
x = np.linspace(0, 6*2*np.pi, num = 1000)
signal_1 = np.sin(2*x)
signal_2 = np.cos(3*x)
signal = signal_1 + signal_2
```

## Plotting signal

```python
fig, ax = plt.subplots(figsize = (16,6))

ax.plot(x,signal)

ax.set_xlim(0, 6*2*np.pi);
```

## FFT in signal

```python
signal_frequency = FFT(signal)
```

## Plotting fequency of signal

```python
fig, ax = plt.subplots(figsize = (16,6))

ax.plot(abs(signal_frequency))

ax.set_xlim(0, 1000);
```

## FFT inverse in signal

```python
signal_again = IFFT(signal_frequency)
```

## Plotting the reconstruct signal

```python
fig, ax = plt.subplots(figsize = (16,6))

ax.plot(x,abs(signal_again))

ax.set_xlim(0, 6*2*np.pi);
```

## Read image

```python
img = mpimg.imread('Images/cameraman.tif')
```

## Plotting image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## FFT image

```python
%%time
img_FFT = FFT2D(img)
```

## Plotting Frequency domain image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(np.power(abs(img_FFT), 1/4), cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## FFT image inverse

```python
%%time
img_again = IFFT2D(img_FFT)
```

## Plotting image reconstruct

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(np.rot90(abs(img_again),2), cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## Plotting images side by side

```python
fig, axs = plt.subplots(1,3, figsize = (18,6))

axs[0].imshow(img, cmap = 'gray')
axs[1].imshow(np.power(abs(img_FFT), 1/4), cmap = 'gray')
axs[2].imshow(np.rot90(abs(img_again),2), cmap = 'gray')

titles = ['(a)\n', '(b)\n', '(c)\n']

for ax, title in zip(axs,titles):
        
    # Set title
        
    ax.set_title(title, fontsize = 24)
        
    # Remove axis
        
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/FFT_side_by_side.png", 
            bbox_inches = 'tight')
```

## FFT scipy

```python
img_fft_spy = fft2(img)
```

## Plotting frequency domain

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(np.power(abs(img_fft_spy), 1/4).astype(np.int32), cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## Compare the frequency domain get of FFT implementation and scipy

```python
fig, axs = plt.subplots(1,2, figsize = (18,6))

axs[0].imshow(np.power(abs(img_FFT), 1/4), cmap = 'gray')
axs[1].imshow(np.power(abs(img_fft_spy), 1/4).astype(np.int32), cmap = 'gray')

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs,titles):
        
    # Set title
        
    ax.set_title(title, fontsize = 24)
        
    # Remove axis
        
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/frequency_domain_implementation_and_scipy.png", 
            bbox_inches = 'tight')
```

## FFT inverse scipy

```python
img_spy = ifft2(img_fft_spy)
```

## Plotting image construct with scipy

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(abs(img_spy), cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```
