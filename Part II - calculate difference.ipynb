## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
import matplotlib.image as mpimg
```

## Read images

```python
komodo = dict()

komodo['original'] = mpimg.imread('Images/komodo.jpg')

for i in range(1,4):
    komodo['composition' + str(i)] = mpimg.imread(
        'Results/komodo_composition' + str(i) + '.jpg')
```

```python
castle = dict()

castle['original'] = mpimg.imread('Images/castle.jpg')

for i in range(1,4):
    castle['composition' + str(i)] = mpimg.imread(
        'Results/castle_composition' + str(i) + '.jpg')
```

## Plotting images

```python
fig, axs = plt.subplots(1,2, figsize = (12,6))

axs[0].imshow(komodo['original'])
axs[1].imshow(castle['original'])

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs,titles):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
```

## Calculate difference

```python
for i in range(1,4):
    komodo['difference' + str(i)] = (100*abs(
        komodo['original'] - komodo['composition' + str(i)])).astype(
        np.int32)
```

```python
for i in range(1,4):
    castle['difference' + str(i)] = (100*abs(
        castle['original'] - castle['composition' + str(i)])).astype(
        np.int32)
```

## Plotting differences

```python
fig, axs = plt.subplots(3,2, figsize = (12,18))

axs[0][0].imshow(komodo['composition1'])
axs[0][1].imshow(komodo['difference1'])
axs[1][0].imshow(komodo['composition2'])
axs[1][1].imshow(komodo['difference2'])
axs[2][0].imshow(komodo['composition3'])
axs[2][1].imshow(komodo['difference3'])

titles = [['(a)', '(b)'], 
          ['(c)', '(d)'], 
          ['(e)', '(f)']]

for axss, titless in zip(axs,titles):
    for ax, title in zip(axss,titless):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/komodo_composition_diference.png", 
            bbox_inches = 'tight')
```

```python
fig, axs = plt.subplots(3,2, figsize = (12,18))

axs[0][0].imshow(castle['composition1'])
axs[0][1].imshow(castle['difference1'])
axs[1][0].imshow(castle['composition2'])
axs[1][1].imshow(castle['difference2'])
axs[2][0].imshow(castle['composition3'])
axs[2][1].imshow(castle['difference3'])

titles = [['(a)', '(b)'], 
          ['(c)', '(d)'], 
          ['(e)', '(f)']]

for axss, titless in zip(axs,titles):
    for ax, title in zip(axss,titless):
        
        # Set title
        
        ax.set_title(title, fontsize = 24)
        
        # Remove axis
        
        ax.xaxis.set_major_locator(NullLocator())
        ax.yaxis.set_major_locator(NullLocator())
        
plt.savefig("Results/castle_composition_diference.png", 
            bbox_inches = 'tight')
```
